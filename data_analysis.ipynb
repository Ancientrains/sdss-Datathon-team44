```python
import pgeocode 
postal_col = "LOCATION_POSTAL_CODE"

postal_code = (
    df[postal_col].astype(str).str.strip().str.upper().replace({"NAN":pd.NA, "":pd.NA})
)
lookup_postal_code = postal_code.str.replace(" ", "", regex=False).str[:3]
geo_lookup = pgeocode.Nominatim("ca").query_postal_code(
    lookup_postal_code.dropna().unique().tolist()
)
geo_lookup = geo_lookup[["postal_code", "latitude", "longitude"]].rename(
    columns={
        "postal_code": "lookup_postal_code",
        "latitude": "lat",
        "longitude": "lon",
    }
)

df = (
    df.assign(lookup_postal_code=lookup_postal_code)
      .merge(geo_lookup, on="lookup_postal_code", how="left")
      .drop(columns=["lookup_postal_code"])
)
print(df[[postal_col, "lat", "lon"]].head(10))
```

```python
from sklearn.metrics.pairwise import haversine_distances
import numpy as np

# --- Gaussian Pressure Index (memory-efficient batched version) ---
# Avoids building the full (n x n) distance matrix by processing BATCH_SIZE
# rows at a time, keeping peak memory to ~(BATCH_SIZE x n) floats.

RADIUS_MILES = 2
SIGMA_MILES  = RADIUS_MILES / 2
RADIUS_RAD   = RADIUS_MILES / 3958.8
SIGMA_RAD    = SIGMA_MILES  / 3958.8
BATCH_SIZE   = 500          # tune up/down to balance speed vs memory

mask  = df["lat"].notna() & df["lon"].notna() & df["OCCUPANCY_RATE"].notna()
valid = df[mask].copy().reset_index()

coords_rad = np.radians(valid[["lat", "lon"]].values)   # (n, 2)
occ        = valid["OCCUPANCY_RATE"].values              # (n,)
n          = len(valid)

pressure_scores = np.empty(n, dtype=np.float64)

for start in range(0, n, BATCH_SIZE):
    end        = min(start + BATCH_SIZE, n)
    batch_rad  = coords_rad[start:end]                   # (b, 2)

    # distances from this batch to ALL points: (b, n)
    dist_chunk = haversine_distances(batch_rad, coords_rad)

    # Gaussian weights, zero beyond radius
    w = np.exp(-0.5 * (dist_chunk / SIGMA_RAD) ** 2)
    w[dist_chunk > RADIUS_RAD] = 0.0

    # Weighted average for each row in the batch
    pressure_scores[start:end] = (w @ occ) / w.sum(axis=1)

    if (start // BATCH_SIZE) % 20 == 0:
        print(f"  processed {end}/{n} rows...")

valid["PRESSURE_SCORE_GAUSSIAN"] = pressure_scores

df["PRESSURE_SCORE_GAUSSIAN"] = np.nan
df.loc[valid["index"], "PRESSURE_SCORE_GAUSSIAN"] = valid["PRESSURE_SCORE_GAUSSIAN"].values

print("Done.")
print(df[["lat", "lon", "OCCUPANCY_RATE", "PRESSURE_SCORE_GAUSSIAN"]].dropna().head(10))
```

```python
import matplotlib.pyplot as plt

pressure = df["PRESSURE_SCORE_GAUSSIAN"].dropna().sort_values().to_numpy()
cdf = np.arange(1, len(pressure) + 1) / len(pressure)

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(pressure, cdf, linewidth=2)
ax.set_title("CDF of PRESSURE_SCORE_GAUSSIAN")
ax.set_xlabel("PRESSURE_SCORE_GAUSSIAN")
ax.set_ylabel("Cumulative probability")
ax.grid(True, alpha=0.3)
plt.show()

print(f"Plotted {len(pressure):,} non-null points.")
```
